Зчитати завантажені текстові файли у фрейм
(https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html) (детальніше
про роботу із фреймами буде розказано у подальших лабораторних роботах).
Імена стовбців фрейму мають бути змістовними та легкими для сприйняття (не
повинно бути спеціалізованих символів, пробілів тощо). Ця задача має бути
реалізована у вигляді окремої процедури, яка на вхід приймає шлях до
директорії, в якій зберігаються файли;

```python
def frame(path):
    headers = ['Year', 'Week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI', 'empty']
    dataframes = []
    
    for filename in os.listdir(path):
        if not filename.endswith('.csv'):
            continue
        parts = filename.split("_")
        if parts[0] != "NOAA":
            continue
        region_id = parts[2]
        filepath = os.path.join(path, filename)
            
        df = pd.read_csv(filepath, header=1, names=headers)
        df = df.drop(df.loc[df['VHI'] == -1].index)
        df['Region'] = region_id
        dataframes.append(df)
        #print(region_id)
    combined_df = pd.concat(dataframes, ignore_index=True).drop_duplicates().reset_index(drop=True)
    combined_df['Year'] = combined_df['Year'].str.replace('<tt><pre>', '').str.replace('</pre></tt>', '')
    combined_df = combined_df[combined_df['Year'].astype(str).str.strip() != '']
    combined_df['Year'] = combined_df['Year'].astype(int)
    combined_df['Region'] = combined_df['Region'].astype(int)
    combined_df = combined_df.drop(columns=['empty'])
    return combined_df

frame('data')

```
